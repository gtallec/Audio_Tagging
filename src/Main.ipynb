### 1. Feature Extraction

```python
chunk_location = data_path + 'chunks/'
sampling_freq_extension = '.16kHz'
file_format_extension = '.wav'

"""SAND BOX
(rate, sig) = scipy.io.wavfile.read(chunk_location + dev_chunks_refined['Chunk'][0] + sampling_freq_extension + file_format_extension)

mfcc_coefficients = mfcc(signal=sig,
                         samplerate=rate,
                         winlen=0.020,
                         winstep=0.010,
                         numcep=24)
mfb_coefficients = logfbank(signal = sig,
                            samplerate=16000,
                            winlen=0.020,
                            winstep=0.010,
                            nfilt=40,
                            nfft=512,
                            lowfreq=0,
                            highfreq=None,
                            preemph=0.97)
"""
def compute_mfb_coef(chunk):
    signal = scipy.io.wavfile.read(chunk_location + chunk + sampling_freq_extension + file_format_extension)
    return logfbank(signal = sig,
                    samplerate=16000,
                    winlen=0.020,
                    winstep=0.010,
                    nfilt=40,
                    nfft=512,
                    lowfreq=0,
                    highfreq=None,
                    preemph=0.97)

dev_chunks_refined['mfb'] = dev_chunks_refined['chunk'].map(lambda chunk : compute_mfb_coef(chunk))
```

```python
def reshape_to_odd(mat, odd):
    n,m = mat.shape
    l = n//odd
    return mat.reshape((l,odd,m))

dev_chunks_refined['mfb'] = dev_chunks_refined['mfb'].map(lambda mfb: reshape_to_odd(mfb, 7))
```
